```python
import matplotlib.pyplot as plt
import numpy as np
A =np.array([[2,-2], [3,1]])
v = np.array([1,2])
transform = A @ v
print(transform)
fig, ax = plt.subplots()
ax.quiver(0, 0, v[0], v[1], angles='xy', scale_units='xy',
          scale=1, color='blue', label='original')
ax.quiver(0, 0, transform[0], transform[1], angles='xy',
          scale_units='xy', scale=1, color='red', label='transformed')
ax.set_xlim(-1, 5)
ax.set_ylim(-1, 5)
ax.legend()
ax.grid(True)
plt.title("Linear Transformation")
plt.show()
```

```python
B = np.array([[0, -1],
              [1,  1]])  # this is a 90 degree rotation

# Apply B first, then A
AB = A @ B
print("A @ B =", AB)

# Apply A first, then B
BA = B @ A
print("B @ A =", BA)

# Are they the same?
print("AB == BA?", np.allclose(AB, BA))
```

```python
Det_A = np.linalg.det(A)
Det_B = np.linalg.det(B)
print(round (Det_A, 2))
print(round(Det_B, 2))
```

```python
def Transformation(matrix, title):
    square = np.array([[0,1,1,0,0], [0,0,1,1,0]])
    transformed = matrix @ square
    fig, (ax1, ax2) = plt.subplots(1, 2, figsize = (10,4))
    ax1.plot(square[0], square[1])
    ax1.set_aspect('equal')
    ax2.plot(transformed[0], transformed[1])
    ax2.set_aspect('equal')
    plt.show()
Transformation(A, "Stretched")
Transformation(B, 'Collapsed')
```

```python
A_inv = np.linalg.inv(A)
print(A_inv)
try:
  B_inv = np. linalg.inv(B)
  print (B_inv)
except:
  np.linalg.LinAlgError
  print('Det = 0')
```

```python
A = np.array([[2,1], [1,3]])
eigenvalues, eigenvectors = np.linalg.eig(A)
print(eigenvalues)
print(eigenvectors)
```

```python
v1 = eigenvectors [:,0]
print(v1)
v2 = eigenvectors [:, 1]
print (v2)
```

```python
print( A @ v1)
print(eigenvalues[0] * v1)
print(np.allclose(A @ v1, eigenvalues[0] * v1))
```
